```python
import pandas as pd
import os

# Lista de nombres de archivos Excel
file_names = ["SLM2019 .xlsx", "SLM2020.xlsx", "SLM2021.xlsx", "SLM2022.xlsx", "SLM2023.xlsx"]

# Lista para almacenar los DataFrames de cada archivo
dfs = []

# Leer cada archivo Excel y cada hoja
for file_name in file_names:
    xls = pd.ExcelFile(file_name)
    for sheet_name in xls.sheet_names:
        df = xls.parse(sheet_name)
        df['File'] = file_name  # Agregar una columna para identificar de qué archivo proviene
        df['Sheet'] = sheet_name  # Agregar una columna para identificar de qué hoja proviene
        dfs.append(df)

# Concatenar todos los DataFrames en uno solo
combined_df = pd.concat(dfs, ignore_index=True)

# Eliminar filas con valores nulos en 'Material Description'
combined_df = combined_df.dropna(subset=['Material Description'])

# Convertir 'mes-año' a tipo de dato datetime
combined_df['mes-año'] = pd.to_datetime(combined_df['mes-año'], errors='coerce')

# Agrupar por 'Material Description' y 'mes-año'
grouped_df = combined_df.groupby(['Material Description', 'mes-año']).first().reset_index()

columns_to_drop = ['Cons.120/36 months', 'Ddisponent', 'Plant', 'Valuation Class', 'Base Unit of Measure', 'Stock market price(LPP)']  # Reemplaza con los nombres reales de las columnas que deseas eliminar
grouped_df = grouped_df.drop(columns=columns_to_drop)
# Guardar el DataFrame combinado en un nuevo archivo Excel
grouped_df.to_excel("combined_data.xlsx", index=False)

# Mostrar las primeras filas del DataFrame combinado
grouped_df.head(20)
```
